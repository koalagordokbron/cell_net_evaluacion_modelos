### Función de clasificación de clases
1. **Sigmoid**

Es la función que aparece por defecto en el notebook proporcionado. Es más apropiada para cuando una entrada puede pertenecer a varias clases, o para problemas de clasificación binaria.

2. **Softmax**

Una entrada solo pertenece a una clase de las 3 que hay en este caso.
Es la que yo usaría, ya que al tratarse del estado de una célula, no puede estar en varios a la vez. Una célula está sana, es un tumor benigno o un tumor maligno. A mi entender, una célula no puede estar sana y ser un tumor maligno a la vez.
